```python
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL Setup for the Tokopedia search page for "laptop core i7 gaming"
url = 'https://www.tokopedia.com/search?st=&q=laptop%20core%20i7%20gaming&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&navsource='

# Selenium Driver Initialization
driver = webdriver.Chrome()
driver.get(url)
```

```python
# scrape products in range of 3 pages (around 250 products)
# loop to scrape data from the first three pages of the search results (each page contains around 84 products)
list_products = []
for page in range(3):

    # Wait for Page Elements to Load
    WebDriverWait(driver, 5).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '#zeus-root')))
    time.sleep(1)

    # Scrolling Down to Load More Products
    for i in range(25):
        driver.execute_script('window.scrollBy(0, 250)')
        time.sleep(1)

    # Horizontal Scrolling Right
    driver.execute_script('window.scrollBy(50, 0)')
    time.sleep(1)

    # Gets the HTML content
    soup = BeautifulSoup(driver.page_source, "html.parser")
    products = soup.find_all('div',class_='css-5wh65g')
    
    # Extract Products Details
    for product in products:
        # data collection
        data = dict()
        data['product_name'] = product.find(class_='OWkG6oHwAppMn1hIBsC3pQ==').text # if product.find(class_='OWkG6oHwAppMn1hIBsC3pQ==') else 'No name'
        data['price'] = product.find(class_='ELhJqP-Bfiud3i5eBR8NWg==').text # if product.find(class_='ELhJqP-Bfiud3i5eBR8NWg==') else 'No price'
        data['rating'] = product.find(class_='nBBbPk9MrELbIUbobepKbQ==').text if product.find(class_='nBBbPk9MrELbIUbobepKbQ==') else 'no rating'
        data['promo'] = product.find(class_='_5+V0nr2fU+1eyI2rpS0FYw==').text if product.find(class_='_5+V0nr2fU+1eyI2rpS0FYw==') else '-'
        data['sold'] = product.find(class_='eLOomHl6J3IWAcdRU8M08A==').text if product.find(class_='eLOomHl6J3IWAcdRU8M08A==') else '-'
        data['seller'] = product.find(class_='X6c-fdwuofj6zGvLKVUaNQ== -9tiTbQgmU1vCjykywQqvA== flip').text if product.find(class_='X6c-fdwuofj6zGvLKVUaNQ== -9tiTbQgmU1vCjykywQqvA== flip') else 'No name'
        data['location'] = product.find(class_='-9tiTbQgmU1vCjykywQqvA== flip').text if product.find(class_='-9tiTbQgmU1vCjykywQqvA== flip') else 'No location info'
        data['link_image'] = product.select_one('div > img').get('src').replace('.webp?ect=4g','') # if product.select_one('div > img') else 'No image'
        
        list_products.append(data)
        
    time.sleep(1)
    
    # Navigate to the Next Page
    driver.find_element(By.CSS_SELECTOR, 'button[aria-label="Laman berikutnya"]').click()
    time.sleep(1)
```

```python
# create Data Frame
data_products = pd.DataFrame(list_products)
data_products
```
